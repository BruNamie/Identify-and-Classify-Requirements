# Creating datasets for training (to be used in Cross-Validation)

This is a notebook that shows how the datasets were created for the training in Cross-Validation

## Libraries

Importing the libraries needed.

```python
import pandas as pd
import numpy as np
```

## Coding

Setting files names.

```python
app = ['eBay', 'Evernote', 'Facebook', 'Netflix', 'PhotoEditor', 'Spotify', 'Twitter', 'WhatsApp']
```

Loading the dataset

```python
df = pd.read_excel('Dataset_REBERT_labeled.xlsx')
df
```

```python
df = df.drop('Unnamed: 0', axis='columns')
df
```

Creating filtered datasets: getting reviews that doesn't belong to the app selected. One dataset for each Cross-Validation iteration.

```python
for i in range(0, len(app)):
    df_filtered = df.copy(deep=True)
    df_filtered.drop(df_filtered.loc[df_filtered['App'] == app[i]].index, inplace=True)
    df_filtered.to_excel('Dataset_train_' + app[i] + '.xlsx')
```
